```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_lightgcn_static = pd.read_csv(r"results/online_static_lightgcn_metrics.csv")
df_lightgcn_dynamic_fix  = pd.read_csv(r"results/online_dynamic_fixed_lightgcn_metrics.csv")
df_lightgcn_dynamic_retrain = pd.read_csv(r"results/online_dynamic_retrain_lightgcn_metrics.csv")

df_graphsage_static = pd.read_csv(r"results/online_static_graphsage_metrics.csv")
df_graphsage_dynamic_fix  = pd.read_csv(r"results/online_dynamic_fix_graphsage_metrics.csv")
df_graphsage_dynamic_retrain = pd.read_csv(r"results/online_dynamic_retrain_graphsage_metrics.csv")

df_gat_static = pd.read_csv(r"results/online_static_gat_metrics.csv")
df_gat_dynamic_fix  = pd.read_csv(r"results/online_dynamic_fix_gat_metrics.csv")
df_gat_dynamic_retrain = pd.read_csv(r"results/online_dynamic_retrain_gat_metrics.csv")
```

## Lightgcn results

```python
plt.figure(figsize=(8, 5))

plt.plot(df_lightgcn_static["step"], df_lightgcn_static["mean_rating"], label="Static Env")
plt.plot(df_lightgcn_dynamic_fix["step"], df_lightgcn_dynamic_fix["mean_rating"], label="Dynamic Env with Static Model")
plt.plot(df_lightgcn_dynamic_retrain["step"], df_lightgcn_dynamic_retrain["mean_rating"], label="Dynamic Env with Model Retraining Every 100 Epochs")

plt.xlabel("Step")
plt.ylabel("Mean Rating")
plt.title("Lightgcn Online Mean Rating over Steps")
plt.legend()
plt.tight_layout()
plt.show()
```

## Graphsage results

```python
plt.figure(figsize=(8, 5))

plt.plot(df_graphsage_static["step"], df_graphsage_static["mean_rating"], label="Static Env")
plt.plot(df_graphsage_dynamic_fix["step"], df_graphsage_dynamic_fix["mean_rating"], label="Dynamic Env with Static Model")
plt.plot(df_graphsage_dynamic_retrain["step"], df_graphsage_dynamic_retrain["mean_rating"], label="Dynamic Env with Model Retraining Every 100 Epochs")

plt.xlabel("Step")
plt.ylabel("Mean Rating")
plt.title("GraphSage Online Mean Rating over Steps")
plt.legend()
plt.tight_layout()
plt.show()
```
